```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("EPL_20_21.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
#create 2 new columns
df['MinsPerMatch']=(df['Mins']/df['Matches']).astype(int)
df['GoalsPerMatch']=(df['Goals']/df['Matches']).astype(float)
df.head()
```

```python
# Total goals

Total_Goals=df['Goals'].sum()
Total_Goals
```

```python
# Penalty Goals

Total_Penalty_Goals = df['Penalty_Goals'].sum()
Total_Penalty_Goals
```

```python
# Penalty Attempts

Total_Penalty_Attempts = df['Penalty_Attempted'].sum()
Total_Penalty_Attempts
```

```python
# Pie chart for penalties scored vs missed
plt.figure(figsize=(13,7))
Penalty_Not_Scored = Total_Penalty_Attempts - Total_Penalty_Goals
data=[Penalty_Not_Scored,Total_Penalty_Goals]
labels=['Penalties missed','Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(data,labels=labels,colors=color,autopct='%.2f%%')
plt.title('Penalties scored vs missed',fontsize=20)
plt.show()
```

```python
# unique positions

df['Position'].unique()
```

```python
Mci_Pen_Attempts=df[df['Club']=='Manchester City'].Penalty_Attempted.sum()
Mci_Pen_Attempts


```

```python
Mci_Pen_Goals = df[df['Club']=='Manchester City'].Penalty_Goals.sum()
Mci_Pen_Goals
```

```python
Mci_Pen_Missed=Mci_Pen_Attempts-Mci_Pen_Goals
Mci_Pen_Missed
```

```python
# Pie chart for penalties scored vs missed for Manchester city
plt.pie([Mci_Pen_Goals,Mci_Pen_Missed],labels=['Penalties Scored','Penalties Missed'],autopct='%.0f%%')
plt.title('Penalties scored vs missed for Manchester city',fontsize=20)
plt.show()
```

```python
#Players from different nations
np.size((df['Nationality'].unique()))
```

```python
# Most players come from which nation

nationality = df.groupby('Nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar',figsize=(12,6),color=sns.color_palette("magma"))
```

```python
# clubs with maximum players in the squad
df['Club'].value_counts().nlargest(5).plot(kind='bar',figsize=(12,5),color=sns.color_palette("viridis"))
                                           
```

```python
# clubs with the least no of players
df['Club'].value_counts().nsmallest(5).plot(kind='bar',figsize=(12,5),color=sns.color_palette("viridis"))
```

```python
#Players based on age group

Under20 = df[df['Age']<=20]
age20_25=df[(df['Age']>20) & (df['Age']<=25)]
age25_30=df[(df['Age']>25) & (df['Age']<=30)]
Above30=df[df['Age']>30]
```

```python
x=np.array([Under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),Above30['Name'].count()])
mylabels=["<=20",">20 & <=25",">25 & <=30",">30"]
plt.title('Total Players with age group',fontsize=20)
plt.pie(x,labels=mylabels,autopct="%.1f%%")
plt.show()
```

```python
# Total under 20 players in each club

Under20['Club'].value_counts().plot(kind='bar',color=sns.color_palette("cubehelix"))
```

```python
# under 20 in ManU
Under20[Under20['Club']=='Manchester United']
```

```python
#average age of players in each club

plt.figure(figsize=(12,6))
sns.boxplot(x='Club',y='Age',data=df)
plt.xticks(rotation = 90)
```

```python
num_player  =df.groupby('Club').size()
data=(df.groupby('Club')["Age"].sum())/num_player
data.sort_values(ascending=False)
```

```python
#Total Assists from each club

Assists_by_clubs=pd.DataFrame(df.groupby('Club', as_index=False)['Assists'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='Club',y='Assists',data=Assists_by_clubs.sort_values(by='Assists'),palette='Set2')
ax.set_xlabel('Club',fontsize=30)
ax.set_ylabel('Assists',fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of clubs vs Total assists',fontsize=25)
```

```python
# Top 10 assists
top_10_assists = df[['Name','Club','Assists','Matches']].nlargest(n=10,columns='Assists')
top_10_assists
```

```python
Goals_by_clubs=pd.DataFrame(df.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='Club',y='Goals',data=Goals_by_clubs.sort_values(by='Goals'),palette='rocket')
ax.set_xlabel('Club',fontsize=30)
ax.set_ylabel('Goals',fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Goals',fontsize=25)
```

```python
# most goals by players 
top_10_goals = df[['Name','Club','Goals','Matches']].nlargest(n=10,columns='Goals')
top_10_goals
```

```python
#Pie chart - Goals with assists and without assist

plt.figure(figsize=(14,7))
assists = df['Assists'].sum()
data=[Total_Goals-assists,assists]
labels=['Goals without assists','Goals with assists']
color=sns.color_palette('Set1')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()
```
